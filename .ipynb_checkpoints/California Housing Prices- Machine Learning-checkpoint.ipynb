```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read csv file
data=pd.read_csv('housing.csv')
```

```python
#preview file headers
data.head()
```

```python
#preview data structures and row numbers
data.info()
```

```python
#drop null values to have cleaner data
data.dropna(inplace=True)
```

```python
#confirm null rows are dropped
data.info()
```

```python
#Split data for train and testing without column being tested
from sklearn.model_selection import train_test_split

X= data.drop(['median_house_value'], axis=1)
y= data['median_house_value']
```

```python
#confirm necessary column is excluded
X
```

```python
#confirm only column being tested
y
```

```python
#Train 80% of data and test 20%
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2)
```

```python
#join train data
train_data= X_train.join(y_train)
```

```python
#confirm the join was done including excluded column
train_data
```

```python
#histrograms to show relations
train_data.hist(figsize=(15, 10))
```

```python
#calculate corrections between columns 
train_data.corr()
```

```python
#use heatmap to better visualize correlations
plt.figure(figsize=(15, 10))
sns.heatmap(train_data.corr(), annot=True, cmap="YlOrBr")
```

```python
#natural logarithm 
train_data['total_rooms']=np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
```

```python
#visualize
train_data.hist(figsize=(15,10))
```

```python
#investigate ocean_priximity column due to its format of words
train_data.ocean_proximity.value_counts()
```

```python
#use get_dummies to convert ocean proximity column to binary type and drop orginial column
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)
```

```python
#confirm ocean proximity is dropped and replaced with dummies
train_data
```

```python
#view heatmap with new correlations
plt.figure(figsize=(15, 10))
sns.heatmap(train_data.corr(), annot=True, cmap="YlOrBr")
```

```python
#use scatterplot to show correlation of area to house value
plt.figure(figsize=(15,10))
sns.scatterplot(x='latitude', y='longitude', data=train_data, hue='median_house_value', palette='coolwarm')
```

```python
#add columns to better engineer correlations
train_data['bedroom_ratio']= train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']= train_data['total_rooms']/train_data['households']
```

```python
#view heatmap with new correlations
plt.figure(figsize=(15, 10))
sns.heatmap(train_data.corr(), annot=True, cmap="YlOrBr")
```

```python
#Use linear reqression model with scaled data
#train data without median house value column
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train, y_train= train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s= scaler.fit_transform(X_train)

reg=LinearRegression()

reg.fit(X_train_s, y_train)
```

```python
#join test and train data
test_data= X_test.join(y_test)

test_data['total_rooms']=np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms']=np.log(test_data['total_bedrooms']+1)
test_data['population']=np.log(test_data['population']+1)
test_data['households']=np.log(test_data['households']+1)

test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

test_data['bedroom_ratio']= test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']= test_data['total_rooms']/test_data['households']

X_test, y_test= test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
#fit test data using scaler
X_test_s= scaler.fit_transform(X_test)
```

```python
#view test data with all modifications
test_data
```

```python
#view train data to confirm all modifications
train_data
```

```python
#run linear regression score
reg.score(X_test_s, y_test)
```

```python
#use Random Forest Regression to attemp better score
from sklearn.ensemble import RandomForestRegressor


forest= RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```

```python
#Use GridSearchCV with parameters to optimize score
from sklearn.model_selection import GridSearchCV

forest= RandomForestRegressor()
param_grid={
    "n_estimators": [100, 200, 300],
    "min_samples_split": [2, 4],
    "max_depth": [None, 4, 8]
}

grid_search= GridSearchCV(forest, param_grid, cv=5,
                         scoring="neg_mean_squared_error",
                         return_train_score=True)
grid_search.fit(X_train_s, y_train)
```

```python
best_forest=grid_search.best_estimator_
```

```python
#Not much improvement from Random Forest Regressor
best_forest.score(X_test_s, y_test)
```
